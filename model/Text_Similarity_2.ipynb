```python
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('all-MiniLM-L6-v2')

#Our sentences we like to encode
sentences = [
    'this flower has a long orange petal with yellow dots on it the flower has a green pedicel with orange petals with red spots on them this flower has six long orange petals with deep red accents this flower is orange and yellow in color, with petals that are spotted this flower has petals that are orange with red dots this flower has long and rounded orange petals with bright red dots this flower has sparsely arranged yellow and orange petals with darker orange spots on them the long think petals on this flower have red spots on them and sit on a green pedicel this flower has petals that are orange and has red spots this flower is orange and yellow in color, and has petals that are spotted.',
              'Yellow Flower with Green Pedicel','Blue Flower with zero petals',
              'this flower has a long orange petal with yellow dots on it the flower has a green pedicel with orange petals with red spots on them'
             ]

#Sentences are encoded by calling model.encode()
embeddings = model.encode(sentences)

#Print the embeddings
for sentence, embedding in zip(sentences, embeddings):
    print("Sentence:", sentence)
    print("")
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
hi = cosine_similarity(embeddings)
print(hi)
```
